# Seasonal Decomposition of Web Traffic

## Objective
Analyze daily website traffic and decompose it into trend, seasonal, and residual components using additive and multiplicative models.

## Dataset
Personal Ecommerce Website â€“ Daily Page Views with Ad Spend, Weekend, and Holiday Indicators.

## Tools Used
- Python
- pandas
- statsmodels
- matplotlib



```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
import pandas as pd
plt.style.use("seaborn-v0_8")

```

## Data Loading
The dataset contains daily page view statistics along with contextual variables such as ad spend, weekends, and holidays.


```python
df = pd.read_csv("final_website_stats3.csv")
df.head()
```

## Data Cleaning
Removing non-informative index columns and preparing the dataset for time-series analysis.


```python
#Drop the useless column
if "Unnamed: 0" in df.columns:
    df = df.drop(columns=["Unnamed: 0"])
```

```python
df["timestamp"] = pd.to_datetime(df["timestamp"])
df = df.set_index("timestamp").sort_index()

# Ensure daily frequency
df = df.asfreq("D")
```

## Additive Seasonal Decomposition
Assumes seasonal effects remain constant over time.


```python
df = df.sort_index()
```

```python
#Confirm daily continuity
df.index.min(), df.index.max(), df.index.freq
```

```python
df = df.asfreq("D")
```

```python
plt.figure(figsize=(14,5))
plt.plot(df.index, df["page_views"], linewidth=1)
plt.title("Daily Website Page Views (Observed)")
plt.xlabel("Date")
plt.ylabel("Page Views")
plt.grid(alpha=0.3)
plt.show()
```

## Additive Seasonal Decomposition
Assumes seasonal effects remain constant over time.


```python
#Additive Decomposition
add_result = seasonal_decompose(df["page_views"], model="additive", period=7)
add_result.plot()
plt.show()
```

## Multiplicative Seasonal Decomposition
Assumes seasonal effects scale with the level of traffic.


```python
#Multiplicative Decomposition
mul_result = seasonal_decompose(df["page_views"], model="multiplicative", period=7)
mul_result.plot()
plt.show()
```

```python
#Compare Additive vs Multiplicative
#Compare residual â€œcleanliness
add_resid_var = add_result.resid.dropna().var()
mul_resid_var = mul_result.resid.dropna().var()
add_resid_var, mul_resid_var
```

### Model Selection
The multiplicative model shows significantly lower residual variance, indicating a better fit for this dataset where seasonal fluctuations scale with traffic volume.


```python
#Residuals (anomalies
resid = add_result.resid.dropna()  # or mul_result.resid.dropna()
top_spikes = resid.sort_values(ascending=False).head(5)
top_drops = resid.sort_values().head(5)

top_spikes, top_drops
```

## Anomaly Analysis
Residual analysis highlights days with unusually high or low traffic. Positive anomalies often align with weekends and high ad spend, while negative anomalies occur on regular weekdays, suggesting potential campaign inefficiencies or external factors.


```python
df.loc[top_spikes.index, ["page_views", "ad_spend", "is_weekend", "is_holiday"]]
df.loc[top_drops.index, ["page_views", "ad_spend", "is_weekend", "is_holiday"]]
```

## Conclusion
This analysis demonstrates how seasonal decomposition can uncover structural patterns in web traffic data. The findings highlight the importance of accounting for multiplicative seasonality in growing platforms and show how residual analysis can be used to identify actionable anomaly candidates for marketing optimization.
